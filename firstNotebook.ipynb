```python
import numpy as np
import pandas as pd
```

```python
#schools data set
schools = pd.read_csv('https://data.cityofnewyork.us/api/views/ig6g-qauv/rows.csv?accessType=DOWNLOAD')
schools.drop(['Principal Name',
              'Principal Title',
              'Principal Email',
              'Principal Phone Number',
              'Fax Number',
              'Superintendent',
              'Superintendent Location Code',
              'Superintendent Email',
              'Community School Sup Name',
              'Community School Sup Email',
              'BFSC Location Code', 
              'BFSC Director Name',
              'BFSC Director Title',
              'BFSC Director Phone',
              'BFSC Director Email',
              'HighSchool Network Location Code',
              'HighSchool Network Name',
              'HighSchool Network Superintendent',
              'HighSchool Network Superintendent Email'],axis=1,inplace=True)
print 'schools dataset size:',schools.shape
schools.head()
```

```python
cols = list(schools.columns)
cols[1] = 'DBN'
cols
schools.columns = cols
schools.head()
cols
```

```python
pupilTeacherRatio = pd.read_csv('https://data.cityofnewyork.us/api/views/rtws-c2ai/rows.csv?accessType=DOWNLOAD')
#merge id schools['ATS System Code'] & pupilTeacherRatio['DBN']
print pupilTeacherRatio.shape
pupilTeacherRatio.head()
```

```python
print schools.shape
print pupilTeacherRatio.shape

schools = pd.merge(schools, pupilTeacherRatio, how='left', on=['DBN', 'DBN'])

print schools.shape

```
